```python
import pandas as pd
from datetime import datetime

# Завантаження даних
df_applications = pd.read_csv('D:\\GO IT\\Python\\ДЗ 5 тема (очистка даних)/applications(2.0).csv')
df_industries = pd.read_csv('D:\\GO IT\\Python\\ДЗ 5 тема (очистка даних)/industries(2.0).csv')

# Об'єднання таблиць за колонкою 'Industry'
df_merged = pd.merge(df_applications, df_industries[['Industry', 'Score']], how='left', on='Industry')

# Заповнення відсутніх значень
df_merged['Score'] = df_merged['Score'].fillna(0)
df_merged['External Rating'] = df_merged['External Rating'].fillna(0)

# Початковий рейтинг
df_merged['Rating'] = 0

# Умови для рейтингу
def calculate_rating(row):
    # Якщо немає 'Amount' або 'External Rating' дорівнює нулю
    if pd.isna(row['Amount']) or row['External Rating'] == 0:
        return 0

    # Ініціалізація рейтингу
    rating = 0

    # Додавання оцінок за критеріями
    # 1. Додавання за віковий діапазон
    if 35 <= row['Age'] <= 55:
        rating += 20

    # 2. Додавання за день подачі заявки (не вихідні)
    submission_date = pd.to_datetime(row['Applied at'], errors='coerce')
    if submission_date is not pd.NaT and submission_date.weekday() < 5:
        rating += 20

    # 3. Додавання за сімейний статус
    if str(row['Marital status']).lower() == 'married':
        rating += 20

    # 4. Додавання за локацію (Київ чи Київська область)
    if str(row['Location']).lower() in ['київ', 'київська область']:
        rating += 10

    # 5. Додавання значення з 'Score' із індустрій
    rating += row['Score']

    # 6. Додавання або віднімання за 'External Rating'
    if row['External Rating'] >= 7:
        rating += 20
    elif row['External Rating'] <= 2:
        rating -= 20

    # Гарантуємо, що рейтинг у межах від 0 до 100
    return max(0, min(100, rating))

# Розрахунок рейтингу для кожного рядка
df_merged['Rating'] = df_merged.apply(calculate_rating, axis=1)

# Збереження результату
df_merged.to_csv('D:\\GO IT\\Python\\ДЗ 5 тема (очистка даних)/applications_final_ratings.csv', index=False)

# Виведення перших кількох рядків для перевірки
print(df_merged[['applicant_id', 'Rating']].head())
```

```python
# Фільтрація заявок із рейтингом > 0
df_accepted = df_merged[df_merged['Rating'] > 0].copy()  # Явно створюємо копію

# Виділення лише дати (до пробілу)
df_accepted['Applied at'] = df_accepted['Applied at'].str.split(' ').str[0]

# Конвертація у формат datetime (лише дата)
df_accepted['Applied at'] = pd.to_datetime(df_accepted['Applied at'], format='%m.%d.%Y', errors='coerce')

# Додавання інформації про тиждень подачі заявки
df_accepted['Submission Week'] = df_accepted['Applied at'].dt.isocalendar().week

# Групування даних за тижнями та обчислення середнього рейтингу
average_rating_by_week = df_accepted.groupby('Submission Week')['Rating'].mean().reset_index()
average_rating_by_week.rename(columns={'Rating': 'Average Rating'}, inplace=True)

# Збереження результатів
df_accepted.to_csv('D:\\GO IT\\Python\\ДЗ 5 тема (очистка даних)/accepted_applications.csv', index=False)
average_rating_by_week.to_csv('D:\\GO IT\\Python\\ДЗ 5 тема (очистка даних)/average_rating_by_week.csv', index=False)

# Виведення результатів
print("Прийняті заявки (з рейтингом > 0):")
print(df_accepted[['applicant_id', 'Rating', 'Submission Week']].head())

print("\nСередній рейтинг прийнятих заявок за тижнями:")
print(average_rating_by_week)
```
